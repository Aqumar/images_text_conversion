```python
# Import libraries 
from PIL import Image 
import pytesseract 
import sys 
from pdf2image import convert_from_path 
import os 
import pandas as pd
import tabula
from tabula import wrapper
  
# Path of the pdf 
PDF_file = "TaxInvoiceKA1192009BD70330.pdf"
  
''' 
Part #1 : Converting PDF to images 
'''
filename = "page_"+str(1)+".jpg"

pages = convert_from_path(PDF_file, 500) 
      
# Save the image of the page in system 
for page in pages:
    page.save(filename, 'JPEG') 

''' 
Part #2 - Recognizing text from the images using OCR 
'''
# Creating a text file to write the output 
outfile = "out_text.txt"
  
# Open the file in append mode so that  
# All contents of all images are added to the same file 
f = open(outfile, "a") 

filename = "page_"+str(1)+".jpg"
          
# Recognize the text as string in image using pytesserct 
text = str(((pytesseract.image_to_string(Image.open(filename))))) 
print(text)  

text = text.replace('-\n', '')     
      
# Finally, write the processed text to the file. 
f.write(text) 

#Close the file after writing all the text. 
f.close() 

df = wrapper.read_pdf(PDF_file, pages = '1', multiple_tables = True)

row_number = df[0].shape[0]-1
column_number = df[0].shape[1]-1

value = list(df[0][row_number:][column_number])[0]
value = value.replace(',','')
value = float(value)
print("Total cost spent on the merchandise is::", value)
```
